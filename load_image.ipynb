```python
# Generate sample data
arr = np.random.rand(3, 4, 5)

# Flatten the array
R_flat = img[:, :, 0].flatten()
G_flat = img[:, :, 1].flatten()
B_flat = img[:, :, 2].flatten()


fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# plot histograms of the flattened values in each subplot
axs[0].hist(R_flat, color='red', bins=50)
axs[0].set_title('Red')
axs[1].hist(G_flat, color='green', bins=50)
axs[1].set_title('Green')
axs[2].hist(B_flat, color='blue', bins=50)
axs[2].set_title('Blue')

# add axis labels and a title to the figure
fig.suptitle('Histogram of RGB Values')
plt.xlabel('Value')
plt.ylabel('Frequency')

# display the figure
plt.show()
```
